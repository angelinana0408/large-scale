```python
import numpy as np
from time import time
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import NullFormatter
from sklearn.manifold import TSNE
import matplotlib.backends.backend_pdf
from collections import OrderedDict
import matplotlib.lines as mlines
import pandas as pd
from ggplot import *


def load_file(file_name):
    
    npArray = np.load(file_name)
    
    return npArray

#choose 3 groups from cora label file
#choose the 3 groups that have largest members
def selectGroups(npDataSet,groupNum):
    
    groupMemberNum = []
    for i in range(npDataSet.shape[1]):
        groupMemberNum.append(0)
    
    #read cora label file to memory, numpy array
    for item in range(npDataSet.shape[0]):
        for groupID in range(npDataSet.shape[1]):
            if npDataSet[item][groupID] == 1:
                groupMemberNum[groupID] += 1
    
    selectedGroupIndex = []
    for j in range(groupNum):
        biggestIndex = groupMemberNum.index(max(groupMemberNum))
        selectedGroupIndex.append(biggestIndex)
        groupMemberNum[biggestIndex] = -1
    
    return selectedGroupIndex
    
    #for x in [1,2]:
        #cora.append([x,x,x,x,x,x,x])
    #cora_group = np.array(cora)
    #print (cora_group)
    #print (cora_group[1])

#get the data of these three group from cora-embedding file
def collectSelectedDataIndex(npDataSet, selectedGroupIndex):
    #print (selectedGroupIndex)
    selectedDataIndex = []
    for dataIndex in range(npDataSet.shape[0]):
        for selectedGroupID in selectedGroupIndex:
            #print(selectedGroupID)
            if npDataSet[dataIndex][selectedGroupID] == 1:
                selectedDataIndex.append(dataIndex)
                break
    return selectedDataIndex

def testMultiLabel(npDataSet):
    
    for item in range(npDataSet.shape[0]):
        labelSum = 0;
        for groupID in range(npDataSet.shape[1]):
            labelSum += npDataSet[item][groupID]
        if labelSum > 1 or labelSum == 0:
            print ("item id: ", item)
    
#reduction the embedding to 2D using T-SNE
def getSelectedEmbeddingData(npDataSet,selectedDataIndex,embedding_file):
    selectedEmbeddingData = []
    npEmbeddingData = load_file(embedding_file)
    for dataIndex in selectedDataIndex:
        selectedEmbeddingData.append(npEmbeddingData[dataIndex])
    npSelectedEmbeddingData = np.array(selectedEmbeddingData)
    return npSelectedEmbeddingData


def tsne(npSelectedEmbeddingData):
    npTsne_embedding = TSNE(n_components=2, perplexity=40,verbose=2, n_iter_without_progress=500).fit_transform(npSelectedEmbeddingData)
    return npTsne_embedding
#visualize the 2D

def drawGraphWithMap(npTsne_embedding, groupSelectedDataMap ):
    style = ['ro','bo', 'go']
    
    styleIndex = 0
    for groupIndex in groupSelectedDataMap:
        for tsneIndex in groupSelectedDataMap[groupIndex]:
            plt.plot(npTsne_embedding[tsneIndex][0],npTsne_embedding[tsneIndex][1],style[styleIndex])
        styleIndex += 1
        
    
    
    plt.show()

#This is for labeled dataset's graph drawing
def drawGraph(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,graphID ):
    style = ['bo', 'go', 'ro', 'co', 'mo', 'yo', 'ko']
    #color = ['red', 'blue', 'green']
    styleIndex = 0
    for groupIndex in selectedGroupIndex:
        for tsneIndex in range(npTsne_embedding.shape[0]):
            dataIndex = selectedDataIndex[tsneIndex]
            if npDataSet[dataIndex][groupIndex] == 1:    
                plt.plot(npTsne_embedding[tsneIndex][0],npTsne_embedding[tsneIndex][1],style[styleIndex], label=str(groupIndex+1) )                
        styleIndex += 1
    plt.title(graphName)
    
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    keylist = by_label.keys()
    by_label_sorted = {}
    for key in sorted(keylist):
        by_label_sorted.update({key: by_label[key]})
    plt.legend(by_label_sorted.values(), by_label_sorted.keys())
    plt.savefig(graphName+"_"+"test"+str(graphID)+".png")
    plt.show()
    
def drawGraph_forUnlabeledDataset(embedding_2d, graphTitle,npLink):
    fig, ax = plt.subplots()
    ax.plot(embedding_2d[:,0],embedding_2d[:,1],'o', ms=15, lw=2, alpha=0.7, mfc='orange')
    for index in range(len(embedding_2d)):
        ax.annotate(str(index+1), xy=(embedding_2d[index][0], embedding_2d[index][1]),ha='center',color='black')
    ax.set_title(graphTitle)
    #######draw link#####
    for link in npLink:
        p1Index = link[0]-1
        p2Index = link[1]-1
        x=[]
        y=[]
        p1 = embedding_2d[p1Index]
        p2 = embedding_2d[p2Index]
        x.append(p1[0])
        x.append(p2[0])
        y.append(p1[1])
        y.append(p2[1])
        ax.plot(x,y,'k-',zorder=-1, linewidth=0.5)
    ########end draw link
    ax.grid()
    plt.show()
    
def plotChart(np2d_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,graphID):
    
    plotSize = 70
    plotAlpha = 0.7
    
    if (np2d_embedding.shape[0] > 3000): 
        plotSize = 30
    
    #npDataSet is groud truth label
    label_number = 0
    for i in range(npDataSet.shape[0]):
        for j in range(npDataSet.shape[1]):
            if i in selectedDataIndex and j in selectedGroupIndex:
                label_number += int(npDataSet[i][j])
    print("label_number ", label_number)
    print("groupIndex ", selectedGroupIndex)
        
    np2d_embedding_withLabel = np.zeros(shape=(label_number,3))
    index_label = 0
    for groupIndex in selectedGroupIndex:
        for index in range(np2d_embedding.shape[0]):
            dataIndex = selectedDataIndex[index]
            if npDataSet[dataIndex][groupIndex] == 1:    
                np2d_embedding_withLabel[index_label][0] = np2d_embedding[index][0]
                np2d_embedding_withLabel[index_label][1] = np2d_embedding[index][1]
                np2d_embedding_withLabel[index_label][2] = groupIndex
                index_label += 1
    
    df_2d = pd.DataFrame(columns=['x','y','label'])
    df_2d['x'] = np2d_embedding_withLabel[:,0]
    df_2d['y'] = np2d_embedding_withLabel[:,1]
    df_2d['label'] = np2d_embedding_withLabel[:,2]
    print("df_2d ", df_2d)
    
    print("np2d_embedding size ", np2d_embedding.shape)
    print("np2d_embedding_withLabel size ", np2d_embedding_withLabel.shape)
    print("df_2d size ", df_2d.shape)

    #df_2d.to_csv('out_'+str(graphName)+str(graphID)+'.csv')
    df_2d['label'] = df_2d['label'].apply(lambda i: str(i))
    
    chart = ggplot(aes(x='x', y='y', color='label'), data=df_2d) + \
        ggtitle(str(graphName)+' (to 200d'+') '+'t-SNE'+' (to 2d'+') ') + \
        geom_point(size=plotSize,alpha=plotAlpha)
        
    
    
    print(chart)
    
    #ggsave(chart, filename = str(graphName)+str(graphID)+".png")

    
def drawGraph_forUnlabeledDataset_networkx(embedding_2d, graphTitle,npLink):
    
    for link in npLink:
        p1Index = link[0]-1
        p2Index = link[1]-1
        x=[]
        y=[]
        p1 = embedding_2d[p1Index]
        p2 = embedding_2d[p2Index]
        x.append(p1[0])
        x.append(p2[0])
        y.append(p1[1])
        y.append(p2[1])
        plt.plot(x,y,'k-',zorder=-1)
    """
    x, y = np.random.random(size=(2,10))
    for i in range(0, len(x), 2):
    plt.plot(x[i:i+2], y[i:i+2], 'ro-')
    """
    plt.show()
    """
    G=nx.Graph()
    G.add_edges_from([(1,2),(2,3),(3,1),(1,4)]) #define G
    fixed_positions = {1:(0,0),2:(-1,2)}#dict with two of the positions set
    fixed_nodes = fixed_positions.keys()
    pos = nx.spring_layout(G,pos=fixed_positions, fixed = fixed_nodes)
    nx.draw(G,pos)
    """
def getGroupSelectedDataMap(npDataSet,selectedGroupIndex, selectedDataIndex):
    
    groupSelectedDataMap = {y: [] for y in selectedGroupIndex}  #initialization
    for groupIndex in selectedGroupIndex:
        for seDataIndex in range(len(selectedDataIndex)):
            dataIndex = selectedDataIndex[seDataIndex]
            if npDataSet[dataIndex][groupIndex] == 1:
                groupSelectedDataMap[groupIndex].append(seDataIndex)
    return groupSelectedDataMap

#######################comment part is for labeled dataset, GCN/N2V/SDNE --> TSNE, using cora dataset################
"""
groupNum = 3
doc_name = "cora/cora_labels.npy"
npDataSet = load_file(doc_name)
selectedGroupIndex = selectGroups(npDataSet,groupNum)
selectedDataIndex = collectSelectedDataIndex(npDataSet, selectedGroupIndex)
groupSelectedDataMap = getGroupSelectedDataMap(npDataSet,selectedGroupIndex, selectedDataIndex)
    
for iterTime in range(0,10):
    
    print("======================test "+str(iterTime+1) +": GCN============================")            

    #test GCN    
    graphName = "GCN--Cora dataset"
    #testMultiLabel(dataSet)
    #embedding_file = "cora/gcn_cora_emb_5000.npy" #this is a 100d embedding
    embedding_file = "cora/gcn_cora_emb_295.npy" #this is a 200d embedding
    npSelectedEmbeddingData = getSelectedEmbeddingData(npDataSet,selectedDataIndex,embedding_file)
    npTsne_embedding = tsne(npSelectedEmbeddingData)
    #print(npTsne_embedding.shape[0])
    #print(npTsne_embedding.shape[1])
    #drawGraphWithMap(npTsne_embedding, groupSelectedDataMap)
    #drawGraph(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,iterTime+1 )
    plotChart(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,iterTime+1 )
    #print ("npTsne_embedding")
    #print (npTsne_embedding)
    
        
    print("======================test "+str(iterTime+1) +": n2v============================") 

    #test node2vec
    graphName = "node2vec--Cora dataset" 
    embedding_file = "cora/cora_emb_iter_1_p_1_q_1_walk_40_win_10.npy" #this is a 200d embedding
    npSelectedEmbeddingData = getSelectedEmbeddingData(npDataSet,selectedDataIndex,embedding_file)
    npTsne_embedding = tsne(npSelectedEmbeddingData)
    #print(npTsne_embedding.shape[0])
    #print(npTsne_embedding.shape[1])
    #drawGraphWithMap(npTsne_embedding, groupSelectedDataMap)
    #drawGraph(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,iterTime+1)
    plotChart(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,iterTime+1 )
    #print ("npTsne_embedding")
    #print (npTsne_embedding)
    
    
    print("======================test "+str(iterTime+1) +": SDNE============================") 

    #test node2vec
    graphName = "SDNE--Cora dataset" 
    embedding_file = "cora/sdne-cora_200d_embedding.npy" #this is a 200d embedding
    npSelectedEmbeddingData = getSelectedEmbeddingData(npDataSet,selectedDataIndex,embedding_file)
    npTsne_embedding = tsne(npSelectedEmbeddingData)
    #print(npTsne_embedding.shape[0])
    #print(npTsne_embedding.shape[1])
    #drawGraphWithMap(npTsne_embedding, groupSelectedDataMap)
    #drawGraph(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,iterTime+1)
    plotChart(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,iterTime+1 )
    #print ("npTsne_embedding")
    #print (npTsne_embedding)

"""

    ########################This part is for labeled dataset, SDNE --> TSNE, using Blog dataset################ 

groupNum = 3
doc_name = "blog/blog_labels.npy"
npDataSet = load_file(doc_name)
selectedGroupIndex = selectGroups(npDataSet,groupNum)
selectedDataIndex = collectSelectedDataIndex(npDataSet, selectedGroupIndex)
groupSelectedDataMap = getGroupSelectedDataMap(npDataSet,selectedGroupIndex, selectedDataIndex)

for iterTime in range(0,10):

    print("======================test "+str(iterTime+1) +": SDNE============================") 

    #test node2vec
    graphName = "SDNE--Blog dataset" 
    embedding_file = "blog/blog_200d_embedding.npy" #this is a 200d embedding
    npSelectedEmbeddingData = getSelectedEmbeddingData(npDataSet,selectedDataIndex,embedding_file)
    npTsne_embedding = tsne(npSelectedEmbeddingData)
    #print(npTsne_embedding.shape[0])
    #print(npTsne_embedding.shape[1])
    #drawGraphWithMap(npTsne_embedding, groupSelectedDataMap)
    #drawGraph(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,iterTime+1)
    plotChart(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,iterTime+1 )
    #print ("npTsne_embedding")
    #print (npTsne_embedding)
    
    
    print("======================test "+str(iterTime+1) +": LINE============================") 

    #test node2vec
    graphName = "LINE--Blog dataset" 
    embedding_file = "blog/line_blog.npy" #this is a 200d embedding
    npSelectedEmbeddingData = getSelectedEmbeddingData(npDataSet,selectedDataIndex,embedding_file)
    npTsne_embedding = tsne(npSelectedEmbeddingData)
    #print(npTsne_embedding.shape[0])
    #print(npTsne_embedding.shape[1])
    #drawGraphWithMap(npTsne_embedding, groupSelectedDataMap)
    #drawGraph(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,iterTime+1)
    plotChart(npTsne_embedding, selectedGroupIndex,selectedDataIndex,npDataSet,graphName,iterTime+1 )
    #print ("npTsne_embedding")
    #print (npTsne_embedding)
   
"""
########################This part is for unlabeled dataset, N2V --> TSNE, using Karate dataset################ 

linkFilename = "karate/Karate_adj.npy"
npLink = load_file(linkFilename)
###########just TSNE(to 2d)
graphName = "TSNE(to 2d)"
embedding_file = "karate/karate_embedding_plain_128.npy"
npEmbDataSet = load_file(embedding_file)
npTsne_embedding = tsne(npEmbDataSet)
drawGraph_forUnlabeledDataset(npTsne_embedding, graphName,npLink )

##########node2vec(to 128d) with TSNE(to 2d)
graphName = "node2vec(to 128d) with TSNE(to 2d)"
embedding_file = "karate/karate_embedding_plain_128.npy"
npEmbDataSet = load_file(embedding_file)
npTsne_embedding = tsne(npEmbDataSet)
drawGraph_forUnlabeledDataset(npTsne_embedding, graphName,npLink )

##########node2vec(to 2d) without TSNE
graphName = "node2vec(to 2d) without TSNE"
embedding_file = "karate/karate_embedding_plain_2.npy"
npEmbDataSet = load_file(embedding_file)
drawGraph_forUnlabeledDataset(npEmbDataSet, graphName,npLink)



#pdf = matplotlib.backends.backend_pdf.PdfPages("output.pdf")
#figure = plt.gcf()
#for fig in figure: ## will open an empty extra figure :(
#    pdf.savefig( fig )
#pdf.close()
"""
```
